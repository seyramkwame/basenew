```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_csv("./DATA/HR-Employee-Attrition.csv")
```

```python
df.head(3)
```

```python
#Beakdown of distance from home by job and attrition
df.groupby(['JobRole', 'Attrition'])['DistanceFromHome'].mean()
```

```python
#Average monthly income by education and attrition
df.groupby(['Education', 'Attrition'])['MonthlyIncome'].mean()
```

```python
#Assignment 3
sns.countplot(x='JobRole', hue='Attrition', data=df)
```

```python
sns.boxplot(x='Education', y='MonthlyIncome', hue='Attrition', data=df)
```

```python
plt.hist(df[df['Attrition'] == 'Yes']['Age'], alpha=0.5, label='Attrition = Yes')
plt.hist(df[df['Attrition'] == 'No']['Age'], alpha=0.5, label='Attrition = No')
plt.legend(loc='upper right')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
